```python
#step1 conenct to api and return json
import requests
import pandas as pd
import json
def _weather():
    res = requests.get(
          "https://community-open-weather-map.p.rapidapi.com/forecast?q="+"London,uk",#json link
          headers={
          "X-RapidAPI-Host": "community-open-weather-map.p.rapidapi.com",
          "X-RapidAPI-Key": "75baeffbb7msha9f983de8fc9710p1ae633jsnd4db8acb65a9"#key from rapid key
        },
  )
    data = res.json()
    return data
_weather() # this is dictionary type output
```

```python
data_json = _weather()
```

```python
df1= pd.json_normalize(data_json['list'],max_level = 5) # dive 5 level down
df2=pd.json_normalize(data_json['list'],'weather',max_level = 5) # weather is nested field so normalize it separately
data_df=df1.drop('weather', axis=1).join(df2)
```

```python
data_df
```

```python
from sqlalchemy import create_engine
engine = create_engine('sqlite:///mydb.db', echo=True)# this creates a new db in the python folder
```

```python
data_df.to_sql('weather_t',engine,if_exists='replace')# replace if exists
```
